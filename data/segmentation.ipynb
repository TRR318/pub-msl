```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
#https://www.kaggle.com/datasets/vetrirah/customer?select=Train.csv
# use only the training data
df = pd.read_csv("data/segmentation_raw.csv", index_col=0)
df
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("gender", OrdinalEncoder(categories=[["Female", "Male"]]), ["Gender"]),
        ("married", OrdinalEncoder(categories=[["No", "Yes"]]), ["Ever_Married"]),
        ("graduated", OrdinalEncoder(categories=[["No", "Yes"]]), ["Graduated"]),
        ("profession", OneHotEncoder(max_categories=3), ["Profession"]),
        ("spending_score", OrdinalEncoder(categories=[["Low", "Average", "High"]]), ["Spending_Score"]),
        ("var_1", OneHotEncoder(max_categories=3), ["Var_1"]),
    ], remainder="passthrough"
)

# Transform the data
transformed_data = preprocessor.fit_transform(df.dropna())

# Convert to DataFrame
feature_names = preprocessor.get_feature_names_out()
df_transformed = pd.DataFrame(transformed_data, columns=feature_names)
df_transformed = df_transformed.iloc[:,~df_transformed.columns.str.endswith('sklearn')]
df_transformed = df_transformed.iloc[:,~df_transformed.columns.str.startswith('var_1')]
df_transformed
```

```python
df_transformed.to_csv("segmentation",index=False)
```
